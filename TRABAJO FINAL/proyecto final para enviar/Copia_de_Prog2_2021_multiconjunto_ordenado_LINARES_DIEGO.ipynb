Operaciones de conjuntos
* Unión.- La unión de dos conjuntos A y B, denotado por  A ? B ; es el conjunto de todos loselementos que están en A, en B o en ambos.
* Intersección.- La intersección de dos conjuntos A y B, denotado por  A ? B; es el conjunto detodos los elementos que pertenecen a ambos A y B simultáneamente.
* Diferencia.- La diferencia de conjuntos A y B, denotado por A - B ; es el conjunto de elementosque pertenecen a A  pero no pertenecen a B
* Pertenencia.- El número entero x es un miembro o pertenece al conjunto A si x es un elementode A.  Esta operación se denota por x ? A .
* Comparaciones.- Los conjuntos pueden compararse entre si mediante el uso de operadoresrelacionales.a.Subconjunto.- El conjunto A es un subconjunto del conjunto B si y solo si todo elemento deA es también un elemento de B.b
* Igualdad.- El conjunto A es igual al conjunto B si y solo si  cada elemento de A está en B ycada elemento de B está en A.c.Desigualdad.- El conjunto A es diferente del conjunto B si y solo si A y B no son iguales.
* Cardinalidad.- La cardinalidad de A, denotado por #A es el número de elementos de A


`````
# Repaso de conjuntos en python
x = {1, 3, 4, 5, 1}
print(x) # {1, 3, 4, 5}

x = set('Hello InDeepLab') # me convierte una lista en un conjunto
print(x) # {'H', 'l', 'o', 'I', 'D', 'p', 'L', 'e', 'n', 'a', ' ', 'b'}

x = {1, 3, 4, 5, 1}
print(x) # {1, 3, 4, 5}
x.add(1)
print(x) # {1, 3, 4, 5}
x.add(17)
print(x) # {1, 3, 4, 5}
x.remove(1)
print(x) # {1, 3, 4, 5}

# Union
{1, 2, 3, 4, 5}.union({3, 4, 5, 6}) # {1, 2, 3, 4, 5, 6}
{1, 2, 3, 4, 5} | {3, 4, 5, 6} # {1, 2, 3, 4, 5, 6}

# Intercepción
{1, 2, 3, 4, 5}.intersection({3, 4, 5, 6}) # {3, 4, 5}
{1, 2, 3, 4, 5} & {3, 4, 5, 6} # {3, 4, 5}

# Diferencia
{1, 2, 3, 4}.difference({2, 3, 5}) # {1, 4}
{1, 2, 3, 4} - {2, 3, 5} # {1, 4}

# Superconjunto
{1, 2}.issuperset({1, 2, 3}) # False
{1, 2} >= {1, 2, 3} # False

{1, 2, 3, 4, 5}.issuperset({1, 2, 3}) # True
{1, 2, 3, 4, 5} >= {1, 2, 3} # True

# Subconjunto
{1, 2}.issubset({1, 2, 3}) # True
{1, 2} <= {1, 2, 3} # True

{1, 5, 3, 4}.issubset({1, 2, 3}) # False
{1, 5, 3, 4} <= {1, 2, 3} # False
`````

Un multiconjunto ordenado es: un multiconjunto (es decir, puede haber repeticiones de elementos) donde los elementos están ordenados. Es decir, un multiconjunto ordenado además de tener indices e indicar elemento 1, 2 , 3 , etc. Tambien cumple que el elemento indice 1 es menor que el elemento 2 y así. 

Se tienen todos los operadores de un multiconjunto,

+ tenemos operadores para los indices, es decir preguntar cual es el elemento en el lugar i y dado un elemento decir en que posición se encuentra. 
+ Por estar ordenado (de menor a mayor) los elementos que esta en el lugar i es menor al elemento i+1

`````
class MulticonjuntoOrdenado():
    datos = []
    def __init__(self):

        """
        Inicializacion del Multiconjunto Ordenado
        """
        self.datos = []
    def agregar(self, elem):
        """
        Agregar un elemento al multiconjunto
        recuerde que debe quedar en orden
        """
        self.datos.append(elem) #(agrega e agrega al final)
        
        for indice in range(1,len(self.datos)):#ordeno por incersion
            valorActual = self.datos[indice]
            posicion = indice
            
            while posicion>0 and self.datos[posicion-1]>valorActual:
                self.datos[posicion]=self.datos[posicion-1]
                posicion = posicion-1

            self.datos[posicion] = valorActual
    def quitar(self, elem):
        """
        Borra una ocurrencia del elemento en el multiconjunto
        """
        corte = 1
        for i in range(len(self.datos)):
            if  corte == 1 and self.datos[i] == elem:
                self.datos.pop(i)
                corte = 0
    def existe(self, elem)->bool:
        """
        Verifica que exista el elemento en el multiconjunto
        """
        primero = 0
        ultimo = len(self.datos)-1
        encontrado = False

        while primero<=ultimo and not encontrado:
            puntoMedio = (primero + ultimo)//2
            if self.datos[puntoMedio] == elem:
                encontrado = True
            else:
                if elem < self.datos[puntoMedio]:
                    ultimo = puntoMedio-1
                else:
                    primero = puntoMedio+1

        return encontrado
    def repeticiones_e(self, elem)->int:
        """
        Cantidad de veces repetidas que aparece el elemento e en el multiconjunto
        """
        contador = 0
        for i in self.datos:
            if elem==i:
                contador = contador +1
            
        return contador
    def primero(self):
        """
        Devuelve el primer elemento del multiconjunto
        """
        if not self.es_vacio():
            return (self.datos[0])
    def ultimo(self):
        """
        Devuelve el ultimo elemento del multiconjunto
        """
        if not self.es_vacio():
            return (self.datos[len(self.datos)-1])
    def devolver_indice(self,elem):
        """
        Devuelve en que orden del multiconjunto se encuentra e
        """
        lista_sin_rep = []
        
        
        for i in self.datos:
            if self.repeticiones_e(i)>=1 and not (i in lista_sin_rep):
                lista_sin_rep.append(i)
                    
        for i in range(len(lista_sin_rep)):
            if lista_sin_rep[i] == elem:
                return i
    def devolver_elemento(self,i:int): 
        """
        Devolveme el elemento que esta en el lugar i
        """
        lista_sin_rep = []
        
        
        for j in self.datos:
            if self.repeticiones_e(j)>=1 and not (j in lista_sin_rep):
                lista_sin_rep.append(j)
                    
        for h in lista_sin_rep:
            if h == lista_sin_rep[i]:
                return h
    def es_vacio(self)->bool:
        """
        Devuelve si el multiconjunto es el vacio
        """
        return self.datos == []
    def tamanio(self)->int:
        """
        Devuelve la cantidad de elementos diferentes del multiconjunto
        """
        l = [] # construyo una lista donde esten todos los elementos sin repeticiones
        for i in self.datos:
            if i not in l:
                l.append(i)
        print(l)
        return len(l)
    def tamanio_rep(self)->int:
        """
        Devuelve la cantidad de elementos del multiconjunto contando repeticiones
        """
        contador = 0
        for i in self.datos:
              contador = contador + 1
        return contador
    def __str__(self):
        """
        Muestra el multiconjunto (hace el print del multiconjunto) (traduccion a string)
        """
        return ("{}".format(self.datos))
`````
